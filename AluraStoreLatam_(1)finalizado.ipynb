### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt # Import the matplotlib library

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

**Categorias de las tiendas**

```python
print(tienda.columns)
print(tienda2.columns)
print(tienda3.columns)
print(tienda4.columns)
```

#1. Análisis de facturación



```python
ingreso_tienda1 = int(tienda['Precio'].sum())
ingreso_tienda2 = int(tienda2['Precio'].sum())
ingreso_tienda3 = int(tienda3['Precio'].sum())
ingreso_tienda4 = int(tienda4['Precio'].sum())

print(f"Ingreso total Tienda 1: {ingreso_tienda1:,} COP")
print(f"Ingreso total Tienda 2: {ingreso_tienda2:,} COP")
print(f"Ingreso total Tienda 3: {ingreso_tienda3:,} COP")
print(f"Ingreso total Tienda 4: {ingreso_tienda4:,} COP")
```

```python
ingresos = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(tiendas, ingresos, color='purple')
plt.title('Ingresos por tienda (COP)')
plt.ylabel('Pesos Colombianos')
plt.xlabel('Tiendas')

plt.show()
```

# 2. Ventas por categoría

```python
tts = [
    (1, tienda),
    (2, tienda2),
    (3, tienda3),
    (4, tienda4)
]

for t in tts:
    print(f"\n📦 Tienda {t[0]} - Ventas por Categoría:\n")

    ventas_por_categoria = (
        t[1].groupby(['Categoría del Producto'])
            .size()
            .sort_values(ascending=False)
            .reset_index(name='Número de Ventas')
    )

    print(ventas_por_categoria)
```

# 5. Envío promedio por tienda

```python
envio_prom_tienda1 = round(tienda["Costo de envío"].mean())
envio_prom_tienda2 = round(tienda2["Costo de envío"].mean())
envio_prom_tienda3 = round(tienda3["Costo de envío"].mean())
envio_prom_tienda4 = round(tienda4["Costo de envío"].mean())

print(f"Costo de envío promedio - Tienda 1: {envio_prom_tienda1} COP")
print(f"Costo de envío promedio - Tienda 2: {envio_prom_tienda2} COP")
print(f"Costo de envío promedio - Tienda 3: {envio_prom_tienda3} COP")
print(f"Costo de envío promedio - Tienda 4: {envio_prom_tienda4} COP")
```

```python
costos_envio = [envio_prom_tienda1, envio_prom_tienda2, envio_prom_tienda3, envio_prom_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.stackplot(tiendas, costos_envio, color="skyblue")
plt.title("Costo de envío promedio por tienda (COP)")
plt.ylabel("Pesos colombianos (COP)")
plt.xlabel("Tiendas")
plt.ylim(0, max(costos_envio) + 500)
plt.tight_layout()
plt.show()
```

## Análisis de todas las tiendas

```python
def analizar_tienda(df, nombre):
    ingresos_totales = df["Precio"].sum()
    promedio_calificacion = df["Calificación"].mean()
    costo_envio_promedio = df["Costo de envío"].mean()
    ventas_por_categoria = df["Categoría del Producto"].value_counts()

    print(f"🔹 {nombre}")
    print(f"Ingresos totales: ${ingresos_totales:,.2f} COP")
    print(f"Valoración media: {promedio_calificacion:.2f}")
    print(f"Costo de envío promedio: ${costo_envio_promedio:.2f} COP")
    print("Ventas por categoría:")
    print(ventas_por_categoria)
    print("-" * 40)

# Ejecutar para cada tienda
analizar_tienda(tienda, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```
